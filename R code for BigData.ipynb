Mini-Project 1: Basic R Skills

```r
# Loading the necessary libraries
library(ggplot2)
library(dplyr)

# Generating the sample data
set.seed(123)
data <- data.frame(
  Category = rep(c("A", "B", "C"), each = 100),
  Value = c(rnorm(100, mean = 50, sd = 10),
            rnorm(100, mean = 60, sd = 15),
            rnorm(100, mean = 70, sd = 20))
)

# Data manipulation
mean_values <- data %>%
  group_by(Category) %>%
  summarise(Average = mean(Value))

# Data visualization: Bar plot of mean values by category
ggplot(mean_values, aes(x = Category, y = Average, fill = Category)) +
  geom_bar(stat = "identity") +
  labs(title = "Mean Values by Category", x = "Category", y = "Average Value") +
  theme_minimal()
```

Mini-Project 2: Advanced R Skills - Fast Data Processing with data.table

```r
library(data.table)

# Generate large sample data
set.seed(123)
large_data <- data.table(
  ID = 1:1000000,
  Value1 = rnorm(1000000),
  Value2 = rnorm(1000000)
)

# Calculate mean for Value1 and Value2 by ID using data.table
result <- large_data[, .(Mean_Value1 = mean(Value1), Mean_Value2 = mean(Value2)), by = ID]

# Display the first few rows of the result
head(result)
```

Mini-Project 3: Visualization and Analysis for Big Data

```r
install.packages("hexbin", repos="http://cran.us.r-project.org")
library(hexbin)
```

```r
# Loading libraries for vizualization
library(ggplot2)
library(hexbin)

# Simulate a large dataset for scatter plot visualization
set.seed(123)
big_data <- data.frame(
  X = rnorm(1000000),
  Y = rnorm(1000000)
)

# Visualization: Hexbin plot for large data
ggplot(big_data, aes(x = X, y = Y)) +
  geom_hex(bins = 50) +
  scale_fill_gradient(low = "lightblue", high = "darkblue") +
  labs(title = "Hexbin Plot of Large Dataset", x = "X", y = "Y") +
  theme_minimal()
```
